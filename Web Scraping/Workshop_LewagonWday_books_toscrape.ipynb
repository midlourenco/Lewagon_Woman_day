```python
import requests
from bs4 import BeautifulSoup
```

```python
url="http://books.toscrape.com/index.html"
response=requests.get(url)
html=response.content
scraped= BeautifulSoup(html,"html.parser")
```

```python
scraped.title #access to the content inside title tag of html code
```

```python
scraped.find("title") #same as above
```

```python
scraped.title.text #it will print invisble characters like \n
```

```python
scraped.title.text.strip() # to remove invisible characters
```

```python
scraped.find("article", class_="product_pod")
```

```python
scraped.article.h3.a #só dá a primeira vez que aparece cada elemento
```

```python
scraped.find_all("article", class_="product_pod") # lista de todos os elementos
```

```python
scraped.find_all("article", class_="product_pod")[2]# we can access to each element of list using the index
#the following is html of 3rd element:
```

```python
items= scraped.find_all("article", class_="product_pod")
for each in items:
    print(each.h3.a["title"])
```

```python
title_links=scraped.find_all("a", title=True)
for each in title_links:
    print(each["title"])
```
